# using matplotlib style for changing bg colour

```python
style.use("ggplot")
plt.bar(classes,class1,width=0.3,align="edge",color="m",edgecolor="y",linewidth=7,alpha=1,label= "class 1 student")
plt.show()
```

# to change the size of graph
# .figure((size))

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1,width=0.3,align="edge",color="m",edgecolor="y",linewidth=7,alpha=1,label= "class 1 student")
plt.show()
```

# to get title and label

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1,width=0.3,align="edge",color="m",edgecolor="y",linewidth=7,alpha=1,label= "class 1 student")
plt.title("bar chart of IAIP class")
plt.xlabel("classes")
plt.ylabel("No of student")
plt.show()
```

# changing the font size of title

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1,width=0.3,align="edge",color="k",edgecolor="y",linewidth=7,alpha=1,label= "class 1 student")
plt.title("bar chart of IAIP class",fontsize=20)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

# two column data in one graph

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
plt.bar(classes,class1,width=0.3,align="edge",color="k",edgecolor="y",linewidth=7,alpha=1,label= "class 1 student")
plt.bar(classes,class2,width=0.3,align="edge",color="m",edgecolor="y",linewidth=7,alpha=1,label= "class 1 student")
plt.title("bar chart of IAIP class",fontsize=20)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.show()
```

# you have to adjust with width 

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
plt.bar(classes_index,class1,width=0.3,align="edge",color="k",edgecolor="y",linewidth=7,alpha=1,label= "class 1 student")
plt.bar(classes_index+0.3,class2,width=0.5,align="edge",color="m",edgecolor="y",linewidth=7,alpha=1,label= "class 2 student")
plt.title("bar chart of IAIP class",fontsize=20)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.legend()
plt.show()
```

# still want give x axis differnt title
# use .xticks( present x lable + width , label you want to add

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
plt.bar(classes_index,class1,width=0.3,align="edge",color="k",edgecolor="y",linewidth=7,alpha=1,label= "class 1 student")
plt.bar(classes_index+0.3,class2,width=0.5,align="edge",color="m",edgecolor="y",linewidth=7,alpha=1,label= "class 2 student")
plt.xticks(classes_index+0.3,classes)
plt.title("bar chart of IAIP class",fontsize=20)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.show()
```

```python
style.use("ggplot")
plt.figure(figsize=(16,9))
classes_index=np.arange(len(classes))
plt.bar(classes_index,class1,width=0.3,align="edge",edgecolor="m",color="y",linewidth=7,alpha=1,label= "class 1 student")
plt.bar(classes_index+0.3,class2,width=0.3,align="edge",edgecolor="k",color="y",linewidth=7,alpha=1,label= "class 2 student")
plt.bar(classes_index+0.3+0.3,class3,width=0.3,align="edge",edgecolor="r",color="y",linewidth=7,alpha=1,label= "class 3 student")
plt.xticks(classes_index+0.3,classes)
plt.title("bar chart of IAIP class",fontsize=20)
plt.xlabel("classes")
plt.ylabel("No of student")
plt.legend()
plt.show()
```
