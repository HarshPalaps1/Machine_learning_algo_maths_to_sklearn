```python
data.dropna(axis=1)
```

# same as second
# drop rows which have any nan values
# functon  .dropna(how="any") 
 

```python
data.dropna(how="any") 
```

# same as thrid
# drop col which have any nan values
# functon  .dropna(how="any") 

```python
data.dropna(axis=1,how="any") 
```

# drop rows which have all nan values
# functon  .dropna(how="all") 


```python
data.dropna(how="all") 
```

# drop col which have all nan values
# functon  .dropna(how="all") 

```python
data.dropna(axis=1,how="all") 
```

# print (row  as default ) which have not null values equal to thresh valoue
# .drop(thresh=number)

```python
data.dropna(thresh=2) 
```

```python
data.dropna(thresh=3) 
```

# print col which have not null values equal to thresh valoue
# .drop(thresh=number)

```python
data.dropna(axis=1,thresh=17)
```

```python
data.dropna(axis=1,thresh=18)
```

# to drop rows of nan values of specific col
