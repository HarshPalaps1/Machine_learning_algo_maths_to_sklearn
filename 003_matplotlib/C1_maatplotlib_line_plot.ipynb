```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
days=np.arange(1,16)
days
```

```python
temp=np.random.randint(30,50,(15))
temp
```

# to create line graph
# .plot(data of x axis,data of y axis)

```python
plt.plot(days,temp)
plt.show()
```

# to create line graph with title
# .plot(data of x axis,data of y axis)
# .title("any char")


```python
plt.plot(days,temp)
plt.title("15 days teamp")
plt.show()
```

# to create line graph with lables
# .plot(data of x axis,data of y axis)
# .title("any char")
# .xlable("any char")
# .ylable("any char")

```python
plt.plot(days,temp)
plt.title("15 days teamp")
plt.xlabel("day number")
plt.ylabel("temp of day")
plt.show()
```

# to create line graph with lables and full axis sclar
# .plot(data of x axis,data of y axis)
# .axis([xmin,xmax,ymin,ymax])
# .title("any char")
# .xlable("any char")
# .ylable("any char")

```python
plt.plot(days,temp)
plt.title("15 days teamp")
plt.axis([0,16,0,50])
plt.xlabel("day number")
plt.ylabel("temp of day")
plt.show()
```

# change the colour of line 
# .plot(x axis ,y axis, colour)

```python
plt.plot(days,temp,"m")
plt.show()
```

# to get different degein for  line
# .plot(x axis ,y axis, colour,marker="any")

```python
plt.plot(days,temp,"r",marker="o")
plt.show()
```

```python
plt.plot(days,temp,"y",marker=">")
plt.show()
```

# to get different style for  line
# .plot(linestyle= "-" or "--" or "-." or ":")

```python
plt.plot(days,temp,"g",marker="^",linestyle="-.")
plt.show()
```

# to get different width 
# .plot(linewidth = num)

```python
plt.plot(days,temp,"b",marker="<",linestyle="--",linewidth=3)
plt.show()
```
