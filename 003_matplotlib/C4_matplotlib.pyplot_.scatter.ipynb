```python
data1=pd.read_csv("D:\\data science\\diploma\\New folder (2)\\mlp\\self\\googleplaystore.csv",nrows=1000)
```

```python
x=data1["Rating"]
y=data1["Reviews"]
```

# to plot scatter plot

```python
plt.scatter(x,y)
plt.show()
```

# to give title and label

```python
plt.scatter(x,y)
plt.title("google play store")
plt.xlabel("rating")
plt.ylabel("Reviws")
plt.show()
```

# 1 to give the color

```python
plt.scatter(x,y,c="y")
plt.title("google play store")
plt.xlabel("rating")
plt.ylabel("Reviws")
plt.show()
```

# 2 to change the size 
# .figure(figsize=(reso))

plt.figure(figsize=(16,9))
plt.scatter(x,y,c="y")
plt.title("google play store")
plt.xlabel("rating")
plt.ylabel("Reviws")
plt.show()

# 3 to change the marker symbol
# marker

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="y",marker="*")
plt.title("google play store")
plt.xlabel("rating")
plt.ylabel("Reviws")
plt.show()
```

# to change maker size
# s

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="y",marker="*",s=1000)
plt.title("google play store")
plt.xlabel("rating")
plt.ylabel("Reviws")
plt.show()
```

# to change the transparency of marker
# alpha

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="y",marker="*",s=1000,alpha=0.1)
plt.title("google play store")
plt.xlabel("rating")
plt.ylabel("Reviws")
plt.show()
```

# to change the edgde color of marker
# edgecolours

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y,c="y",marker="*",s=1000,edgecolors="k")
plt.title("google play store")
plt.xlabel("rating")
plt.ylabel("Reviws")
plt.show()
```

```python

plt.figure(figsize=(16,9))
plt.scatter(x,y,c="y",marker="*",s=1000,linewidths=3,edgecolors="k")
plt.title("google play store")
plt.xlabel("rating")
plt.ylabel("Reviws")
plt.show()
```
