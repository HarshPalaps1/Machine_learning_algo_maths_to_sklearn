```python
import matplotlib.pyplot as plt
import numpy as np
import random 
```

# plot histograms

```python
ml_stu=np.random.randint(18,45,(100))
py_stu=np.random.randint(15,40,(100))
```

```python
print(ml_stu)
```

```python
print(py_stu)
```

# histrogram for ml student

```python
plt.hist(ml_stu)
plt.show()
```

```python
plt.hist(ml_stu)
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.show()
```

# learning .hist() parameters 

# 1 bin to set the range of bar

```python
bins=[18,21,24,27,30,33,36,39,42,45]
plt.hist(ml_stu,bins)
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.show()
```

# 2  to set the rwidth of bar

```python
bins=[18,21,24,27,30,33,36,39,42,45]
plt.hist(ml_stu,bins,rwidth=0.8)
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.show()
```

# 3 histtype { bar,barstacked,step,stepfiled}

```python
bins=[18,21,24,27,30,33,36,39,42,45]
plt.hist(ml_stu,bins,histtype="step")
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.show()
```

# 4 orientation { vertical,horizontal}

```python
bins=[18,21,24,27,30,33,36,39,42,45]
plt.hist(ml_stu,bins,rwidth=0.5,orientation="horizontal")
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.show()
```

# 5 color

```python
bins=[18,21,24,27,30,33,36,39,42,45]
plt.hist(ml_stu,bins,histtype="step",color="y")
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.show()
```

# 6  lable bu .legend("any thing")

```python
bins=[18,21,24,27,30,33,36,39,42,45]
plt.hist(ml_stu,bins,rwidth=0.5,orientation="horizontal",color="m",label="ML student")
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.legend()
plt.show()
```

# two data in one graph

```python
bins=[15,18,21,24,27,30,33,36,39,42,45]
plt.hist(ml_stu,bins,rwidth=0.5,orientation="vertical",color="m",label="ML student")
plt.hist(py_stu,bins,rwidth=0.5,orientation="vertical",color="y",label="py student")
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.legend()
plt.show()
```

```python
bins=[15,18,21,24,27,30,33,36,39,42,45]
plt.hist([ml_stu,py_stu],bins,rwidth=0.5,orientation="vertical",color=["m","y"],label=["ML student","py student"])
#plt.hist(py_stu,bins,rwidth=0.5,orientation="vertical",color="y",label="py student")
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.legend()
plt.show()
```

# to increase the size pf graph

```python
bins=[15,18,21,24,27,30,33,36,39,42,45]
plt.figure(figsize=(16,9))
plt.hist([ml_stu,py_stu],bins,rwidth=0.5,orientation="vertical",color=["m","y"],label=["ML student","py student"])
plt.title("machine learning studend age")
plt.xlabel("number of student")
plt.ylabel("age of the student")
plt.legend()
plt.show()
```
