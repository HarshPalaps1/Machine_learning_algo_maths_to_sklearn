# filling values with scalar number
# .fillna(scalar_number)

```python
data.fillna(0)
```

# filling values of different col with different value with the help dictionary
# .fillna({"col_name":"value"})

```python
data.fillna({"Student id":"col1","Course id":"col2"})
```

# filling values with previous row cell value with method attribute
# .fillna(method="ffill")

```python
data.fillna(method="ffill")
```

# filling values with cell value below it with method attribute
# .fillna(method="bfill")

```python
data.fillna(method="bfill")
```

# filling values with previous col cell with method and axis attribute
# .fillna(method="ffill", axis = 1)

```python
data.fillna(method="ffill",axis=1)
```

# filling values with forword col cell value with method and axis attribute
# .fillna(method="ffill" axis = 0)

```python
data.fillna(method="bfill",axis=1)
```

# filling values with limit and any attribute like value or method
#  .fillna(limit=number of values of each col you want to fill)


```python
data.fillna(0,limit=1)
```

```python
data.fillna(0,limit=3)
```
