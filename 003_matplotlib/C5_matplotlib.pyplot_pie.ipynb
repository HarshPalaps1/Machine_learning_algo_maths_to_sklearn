```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','Machaine Learning','Artificial Intelligence','Data Science']
class1=[45,15,35,25,30]
```

# simplr pie chart

```python
plt.pie([1])
plt.show()
```

# pie chart of above data

```python
plt.pie(class1)
plt.show()
```

# pie with label

```python
plt.pie(class1,labels=classes)
plt.show()
```

# to take one peice out 
# explode

```python
plt.pie(class1,labels=classes,explode=[0.5,0,0,0,0])
plt.show()
```

# to give color to each slice
# by agument colors=["color symbol"]

```python
plt.pie(class1,labels=classes,explode=[0.1,0,0,0,0],colors=["c","b","r","k","m"])
plt.show()
```

# to get percentage of each slice
# by arg autopct = "%o.number of  decimal place you want%%"

```python
plt.pie(class1,labels=classes,autopct="%0.2f%%")
plt.show()
```

# to get shadow in pie turn shadow=True

```python
plt.pie(class1,labels=classes,autopct="%0.2f%%",shadow=True)
plt.show()
```

# to increase the radius of pie 
# arg radius=num

```python
plt.pie(class1,labels=classes,autopct="%0.2f%%",shadow=True,radius=2)
plt.show()
```

```python
# to rotate pie
# arg startangle
```

```python
plt.pie(class1,labels=classes,autopct="%0.2f%%",shadow=True,radius=1,startangle=180)
plt.show()
```

# to change text size
# arg textprops


```python

plt.pie(class1,labels=classes,autopct="%0.2f%%",shadow=True,radius=2,textprops={"fontsize":15})
plt.show()
```
