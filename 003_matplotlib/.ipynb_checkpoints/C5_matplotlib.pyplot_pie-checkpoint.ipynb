```python
import matplotlib.pyplot as plt
```

```python
classes=['Python','R','Machaine Learning','Artificial Intelligence','Data Science']
class1=[45,15,35,25,30]
```

# simplr pie chart

```python
plt.pie([1])
plt.show()
```

# pie chart of above data

```python
plt.pie(class1)
plt.show()
```

# pie with label

```python
plt.pie(class1,labels=classes)
plt.show()
```

# to take one piace out 
# explode

```python
plt.pie(class1,labels=classes,explode=[0.2,0,0,0,0])
plt.show()
```
